```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv('/StudentsPerformance.csv')
```

```python
#1. Calculate the mean, median, and mode for the math, reading, and writing scores
for col in ['math score', 'reading score', 'writing score']:
    mean   = df[col].mean()
    median = df[col].median()
    mode   = df[col].mode()[0]
    print(f"\n{col}")
    print(f"Mean:{mean}")
    print(f"Median:{median}")
    print(f"Mode:{mode}")

```

```python
#2. Determine the range, variance, and standard deviation for each score category.
for col in ['math score', 'reading score', 'writing score']:
    rng = df[col].max() - df[col].min()
    var = df[col].var()
    std = df[col].std()
    print(f"\n{col}")
    print(f"Range:{rng}")
    print(f"Variance:{var}")
    print(f"Std Dev:{std}")
```

```python
#3. Identify the mean, median, and mode of math scores for students with different
#parental education levels
grouped = df.groupby('parental level of education')['math score']
for edu, grp in grouped:
    mean    = grp.mean()
    median  = grp.median()
    mode = grp.mode()[0]
    print(f"\n  {edu}")
    print(f"Mean:{mean}")
    print(f"Median:{median}")
    print(f"Mode:{mode}")
```

```python
#4. Compare the mean scores of students who took the test preparation course versus
#those who didn’t
for col in ['math score', 'reading score', 'writing score']:
    grps = df.groupby('test preparation course')[col].mean()
    print(f"\n{col}")
    for label, val in grps.items():
        print(f"{label}:{val}")
```

```python
#5. Calculate the interquartile range (IQR) for each score category.
for col in ['math score', 'reading score', 'writing score']:
    q1  = df[col].quantile(0.25)
    q3  = df[col].quantile(0.75)
    iqr = q3 - q1
    print(f"\n{col}")
    print(f"Q1:{q1}")
    print(f"Q3:{q3}")
    print(f"IQR:{iqr}")
```

```python
#6. Determine the standard deviation for math scores by lunch type (free/reduced vs.
#standard).
lunch_std = df.groupby('lunch')['math score'].std()
for label, val in lunch_std.items():
    print(f"{label}: {val}")
```

```python
#7. Find the correlation coefficient between math and reading scores, and then between
#reading and writing scores.
math_reading   = df['math score'].corr(df['reading score'])
reading_writing = df['reading score'].corr(df['writing score'])
print(f"Math & Reading: {math_reading}")
print(f"Reading & Writing: {reading_writing}")
```

```python
#8. Interpret the strength and direction of the correlation. What might this suggest about
#the relationship between different scores?
def interpret(r):
    abs_r = abs(r)
    direction = "positive" if r > 0 else "negative"
    if abs_r >= 0.9:
      strength = "very strong"
    elif abs_r >= 0.7:
      strength = "strong"
    elif abs_r >= 0.5:
      strength = "moderate"
    else:
      strength = "weak"
    return f"{strength} {direction}"

print(f"\n Math & Reading({math_reading}):{interpret(math_reading)}")
print(f"\nReading & Writing ({reading_writing}):{interpret(reading_writing)}")

```

```python
#9. Calculate the covariance between math and reading scores. How does it compare
#with the correlation? What does it suggest?
cov = df['math score'].cov(df['reading score'])
print(f"Covariance:{cov}")
print(f"Correlation:{math_reading}")
print("Both are positive suggests scores move together.")
```

```python
#10. Z-Test: Assume the national average math score is 50. Use a one-sample z-test to test
#if this sample's average math score is significantly different from the national
#average

mu0= 50
math_mean = df['math score'].mean()
math_std  = df['math score'].std()
n = len(df)
z_stat= (math_mean - mu0) / (math_std / np.sqrt(n))
p_value= 2 * (1 - stats.norm.cdf(abs(z_stat)))
print(f"Sample mean: {math_mean}")
print(f"Sample std : {math_std}")
print(f"n          : {n}")
print(f"Z-statistic: {z_stat}")
print(f"p-value    : {p_value}")

if p_value < 0.05:
    print(f" The sample mean ({math_mean}) is significantly different from 50.")
else:
    print(f"No significant difference from the national average of 50.")
```

```python
#11. T-Test: Compare the reading scores between students with different lunch types
#(free/reduced vs. standard) to test if there is a significant difference.
free_red  = df[df['lunch']=='free/reduced']['reading score']
standard  = df[df['lunch']=='standard']['reading score']
t_stat, t_pval = stats.ttest_ind(standard, free_red)
print(f"Standard lunch mean: {standard.mean()}(n={len(standard)})")
print(f"Free/reduced lunch mean: {free_red.mean()}(n={len(free_red)})")
print(f"T-statistic: {t_stat}")
print(f"p-value: {t_pval}")
if t_pval < 0.05:
    print("Reading scores differ significantly between lunch groups.")
```

```python
#12. F-Test: Check if the variances of math scores differ significantly between students
#with and without test preparation
completed = df[df['test preparation course']=='completed']['math score']
none= df[df['test preparation course']=='none']['math score']
var_comp = completed.var()
var_none = none.var()
# Larger variance in numerator
if var_comp >= var_none:
    f_stat = var_comp / var_none
    dfn, dfd = len(completed)-1, len(none)-1
    label = "completed / none"
else:
    f_stat = var_none / var_comp
    dfn, dfd = len(none)-1, len(completed)-1
    label = "none / completed"

# Two-tailed p-value
p_upper = 1 - stats.f.cdf(f_stat, dfn, dfd)
f_pval  = 2 * min(p_upper, 1 - p_upper)

print(f"  Variance (completed): {var_comp}(n={len(completed)})")
print(f"  Variance (none): {var_none} (n={len(none)})")
print(f"  F-statistic ({label}) : {f_stat}")
print(f"  df numerator / denom: {dfn} / {dfd}")
print(f"  p-value (two-tailed): {f_pval}")
if f_pval < 0.05:
    print("The variances are significantly different")
else:
    print("No significant difference in variance")
```
