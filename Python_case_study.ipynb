1. Bank Account Manager
a. Simulate a bank account with deposit, withdrawal, and balance
enquiry operations
b. Use variables and data types to store account information
c. Implement conditional statements to check for sufficient balance
before withdrawal
d. Use functions to perform deposit, withdrawal, and balance inquiry
operations.

```python
#1a.Bank account function
def menu():
    print("WELCOME TO PYTHON BANK\n")
    while True:
     print("Select an option:\n 1.Deposit\n 2.Withdraw\n 3.Balance enquiry\n 4.Exit\n")
     choice = input("\nEnter choice (1-4): ")

     if choice == "1":
           amount = float(input("Enter deposit amount: "))
           deposit(account, amount)

     elif choice == "2":
           amount = float(input("Enter withdrawal amount: ₹"))
           withdraw(account, amount)

     elif choice == "3":
            check_balance(account)

     elif choice == "4":
            print("Thank you for banking with us. Goodbye!\n")
            return

     else:
            print("Invalid option. Please choose between 1 and 4.\n")


#1b.Use variables and data types to store account information
account = {
    "account_number": "A12345",   # str
    "holder_name":    "Ann M",  # str
    "balance":        10000.0      # float
}


#1c:Implement conditional check to check sufficient balance before withdrawal
#1d. Use functions to perform deposit, withdrawal, and balance inquiry operations.
def deposit(account, amount):
    account["balance"] += amount
    print(f" Deposited ₹{amount} successfully.")
    print(f"New Balance: ₹{account['balance']}\n")


def withdraw(account, amount):
    # Conditional check
    if amount > account["balance"]:
        print(f"Insufficient balance. Available: {account['balance']}\n")
        return

    account["balance"] -= amount
    print(f"Withdrew {amount} successfully.")
    print(f"New Balance: {account['balance']}\n")


def check_balance(account):
    print(f"  Account Number : {account['account_number']}")
    print(f"  Account Holder : {account['holder_name']}")
    print(f"  Current Balance: {account['balance']}")

menu()

```

2. Weather Data Analyzer
a. Analyze weather data (temperature, humidity, etc.) for a given region
b. Use variables and data types to store weather data
c. Implement conditional statements to determine weather conditions
(sunny, rainy, etc.)
d. Use functions to calculate average temperature and humidity, and print
results


```python
#2a. Analyze weather data for a given region
#2b. Use variables and data types to store weather data
region = "Bengaluru, Karnataka"

weather_data = {
    "monday":    {"temperature": 32.5, "humidity": 78, "wind_speed": 12.0, "rainfall": 0.0},
    "tuesday":   {"temperature": 28.0, "humidity": 85, "wind_speed": 18.5, "rainfall": 12.4},
    "wednesday": {"temperature": 24.5, "humidity": 92, "wind_speed": 22.0, "rainfall": 35.0},
    "thursday":  {"temperature": 26.0, "humidity": 88, "wind_speed": 15.0, "rainfall": 8.0},
    "friday":    {"temperature": 34.0, "humidity": 60, "wind_speed": 8.0,  "rainfall": 0.0},
    "saturday":  {"temperature": 35.5, "humidity": 55, "wind_speed": 6.5,  "rainfall": 0.0},
    "sunday":    {"temperature": 30.0, "humidity": 70, "wind_speed": 10.0, "rainfall": 2.5},
}


#2c.

def get_weather_condition(temperature, humidity, rainfall, wind_speed):
    """
    Determine weather condition using conditional statements.
    Returns a descriptive string for the condition.
    """
    if rainfall > 20:
        condition = "Heavy Rain"
    elif rainfall > 5:
        condition = "Rainy"
    elif rainfall > 0:
        condition = "Light Rainfall"
    elif temperature >= 30 and humidity < 65:
        condition = "Sunny"
    elif temperature >= 26 and humidity >= 75:
        condition = "Humid"
    elif wind_speed > 20:
        condition = "Windy"
    else:
        condition = "Pleasant"

    return condition

print("Please enter day to check weather condition")
day=input()
day = day.strip().lower()

record    = weather_data[day]
condition = get_weather_condition(
        record["temperature"],
        record["humidity"],
        record["rainfall"],
        record["wind_speed"]
    )
print(f"  Weather Report — {day}")
print(f"  Region: {region}")
print(f"  Condition   : {condition}")


 #2d.Average temperature and humidity

def calculate_average_temperature(data):
    total = 0
    for day in data:
        total += data[day]["temperature"]
    average = total / len(data)
    return average


def calculate_average_humidity(data):
   total = 0
   for day in data:
        total += data[day]["humidity"]
   average = total / len(data)
   return average


avg_temp     = calculate_average_temperature(weather_data)
avg_humidity = calculate_average_humidity(weather_data)
print("Average Temperature\n")
print(avg_temp)
print("\n")
print("Average Humidity\n")
print(avg_humidity)

```
