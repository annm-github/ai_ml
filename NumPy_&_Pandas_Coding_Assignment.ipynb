Q1. Employee Salary Adjustment
A company stores salaries in a NumPy array.
salaries = np.array([45000, 52000, 61000, 48000, 75000])
Tasks:
1. Increase all salaries by 10%.
2. Find the average salary.
3. Find the maximum and minimum salary.
4. Print salaries greater than 60000.

```python
import numpy as np
import pandas as pd

# Original salary data
salaries = np.array([45000, 52000, 61000, 48000, 75000])

print("Original Salaries:")
print(salaries)

# Task 1: Increase all salaries by 10%

adjusted_salaries = salaries * 1.10   # multiply each element by 1.10

print("Task 1 — Salaries After 10% Increase:")
print(adjusted_salaries)

# Task 2: Find the average salary (on adjusted salaries)

average_salary = np.mean(adjusted_salaries)

print(f"Task 2 — Average Salary (after raise): Rs.{average_salary}")

# Task 3: Find the maximum and minimum salary

max_salary = np.max(adjusted_salaries)
min_salary = np.min(adjusted_salaries)

print(f"Task 3 — Maximum Salary: Rs.{max_salary}")
print(f"         Minimum Salary: Rs.{min_salary}")

# Task 4: Print salaries greater than 60,000

high_salaries = adjusted_salaries[adjusted_salaries > 60000]

print("Task 4 — Salaries Greater Than Rs.60,000:")
print(high_salaries)

```

Q2. Sales Data Analysis
Daily sales of a store for 10 days:
sales = np.array([120,150,180,200,170,160,210,220,190,175])
Tasks:
1. Compute total sales.
2. Compute mean and standard deviation.
3. Identify days where sales exceeded the average.
4. Normalize the sales using Min-Max scaling.

```python
import numpy as np
import pandas as pd

# Daily sales data for 10 days
sales = np.array([120, 150, 180, 200, 170, 160, 210, 220, 190, 175])
days  = np.arange(1, 11)   # Day 1 through Day 10

# Task 1: Compute total sales

total_sales = np.sum(sales)
print(f"Task 1 — Total Sales: Rs.{total_sales}")

# Task 2: Compute mean and standard deviation

mean_sales = np.mean(sales)
std_sales  = np.std(sales)

print(f"Task 2 — Mean Sales : Rs.{mean_sales}")
print(f"Standard Deviation: Rs.{std_sales}")

# Task 3: Identify days where sales exceeded the average

above_avg_mask = sales > mean_sales          # boolean array
above_avg_days  = days[above_avg_mask]        # days above average
above_avg_sales = sales[above_avg_mask]       # corresponding sales

print(f"Task 3 — Days Where Sales Exceeded Average (Rs.{mean_sales:}):")
for d, s in zip(above_avg_days, above_avg_sales):
    print(f"Day {d} Rs.{s}")

# Task 4: Normalize sales using Min-Max Scaling

sales_min        = np.min(sales)
sales_max        = np.max(sales)
normalized_sales = (sales - sales_min) / (sales_max - sales_min)

print(f"Task 4 — Min-Max Normalized Sales  (Min=Rs.{sales_min}, Max=Rs.{sales_max}):")

for d, s, n in zip(days, sales, normalized_sales):
    print(f"Day {d} sales:Rs.{s}   normalized_sales:{n}")

```

Q3. Matrix Operations for Business Data
A company has sales data for 3 products across 4 regions.
sales_matrix = np.array([
[120,140,130,150],
[80,90,100,95],
[200,210,190,205]
])
Tasks:
1. Find total sales per product.
2. Find total sales per region.
3. Identify the best performing product.
4. Identify the region with highest sales

```python
import numpy as np
import pandas as pd

sales_matrix = np.array([
    [120, 140, 130, 150],
    [ 80,  90, 100,  95],
    [200, 210, 190, 205]])

products = ["Product1", "Product2", "Product3"]
regions  = ["RegionA", "RegionB", "RegionC", "RegionD"]

# Task 1: Total sales per product
total_per_product = np.sum(sales_matrix, axis=1)

print("Task 1 — Total Sales Per Product:")

for product, total in zip(products, total_per_product):
    print(f"  {product}: {total}")

# Task 2: Total sales per region
total_per_region = np.sum(sales_matrix, axis=0)

print("Task 2 — Total Sales Per Region:")
for region, total in zip(regions, total_per_region):
    print(f"  {region}: {total}")

# Task 3: Identify the best performing product
best_product_index = np.argmax(total_per_product)
best_product_name  = products[best_product_index]
best_product_sales = total_per_product[best_product_index]

print("Task 3 — Best Performing Product:")
print(f"  {best_product_name} with total sales = {best_product_sales}")

# Task 4: Identify the region with highest sales
best_region_index = np.argmax(total_per_region)
best_region_name  = regions[best_region_index]
best_region_sales = total_per_region[best_region_index]

print("Task 4 — Region With Highest Sales:")
print(f"{best_region_name}:Total sales = {best_region_sales}")
```

Q4. Employee Dataset Exploration
Create a DataFrame:
data = {
'Employee':['A','B','C','D','E','F'],
'Department':['HR','IT','IT','Finance','HR','Finance'],
'Salary':[50000,65000,70000,72000,52000,68000],
'Experience':[2,5,7,6,3,4]
}
Tasks:
1. Convert this dictionary into a DataFrame.
2. Display basic statistics of the dataset.
3. Find the average salary per department.
4. Identify employees with experience greater than 4 years.
5. Add a new column Bonus = 10% of salary.


```python
import numpy as np
import pandas as pd

# Raw employee data as a dictionary

data = {
    'Employee'  : ['A', 'B', 'C', 'D', 'E', 'F'],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR', 'Finance'],
    'Salary'    : [50000, 65000, 70000, 72000, 52000, 68000],
    'Experience': [2, 5, 7, 6, 3, 4]
}
#Task 1:Convert dictionary into a DataFrame

df = pd.DataFrame(data)
print("Task 1 — Employee DataFrame:")

print(df.to_string(index=False))

#Task 2: Display basic statistics of the dataset

print("Task 2-Basic Statistics:")
print(df.describe())

#Task 3: Average salary per department
avg_salary_dept = df.groupby('Department')['Salary'].mean()

print("Task 3 — Average Salary Per Department:")
for dept, avg in avg_salary_dept.items():
    print(f"  {dept}: Rs.{avg}")
#Task 4: Employees with experience greater than 4 years

experienced_df = df[df['Experience'] > 4]
print("Task 4 — Employees With Experience > 4 Years:")
print(experienced_df.to_string(index=False))

#Task 5: Add a new column Bonus = 10% of Salary
df['Bonus'] = df['Salary'] * 0.10
print("Task 5 — DataFrame With Bonus Column (10% of Salary):")

print(df.to_string(index=False))
```

Q5. Missing Data Handling
Dataset:
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}
Tasks:
1. Detect missing values.
2. Replace missing Age with mean age.
3. Replace missing Salary with median salary.
4. Remove rows with any remaining missing values

```python
import numpy as np
import pandas as pd

# Dataset with missing values
data = {
    'Name'  : ['John', 'Anna', 'Peter', 'Linda', 'David'],
    'Age'   : [28, np.nan, 35, 32, np.nan],
    'Salary': [50000, 54000, np.nan, 58000, 60000]
}

df = pd.DataFrame(data)

# Task 1: Detect missing values
print("Task 1 — Missing Values:")
print(df.isnull())

# Task 2: Replace missing Age with mean age
mean_age    = df['Age'].mean()
df['Age']   = df['Age'].fillna(mean_age)

print(f"Task 2 — Age filled with Mean ({mean_age:}):")
print(df.to_string(index=False))

# Task 3: Replace missing Salary with median salary

median_salary  = df['Salary'].median()
df['Salary']   = df['Salary'].fillna(median_salary)

print(f"Task 3 — Salary filled with Median ({median_salary:}):")
print(df.to_string(index=False))

# Task 4: Remove rows with any remaining missing values

df_clean = df.dropna()

print("Task 4 — After Dropping Remaining NaN Rows:")
print(df_clean.to_string(index=False))
```

Q6. Sales Performance Analysis
Dataset:
sales_data = {
'Region':['North','South','East','West','North','South','East','West'],
'Salesperson':['A','B','C','D','E','F','G','H'],
'Sales':[12000,15000,13000,17000,16000,14000,15500,16500]
}
Tasks:
1. Find total sales per region.
2. Identify the top salesperson.
3. Compute average sales across regions.
4. Sort the dataset by sales descending.


```python
import numpy as np
import pandas as pd

# Sales performance dataset

sales_data = {
    'Region'     : ['North','South','East','West','North','South','East','West'],
    'Salesperson': ['A','B','C','D','E','F','G','H'],
    'Sales'      : [12000,15000,13000,17000,16000,14000,15500,16500]
}

df = pd.DataFrame(sales_data)

# Task 1: Find total sales per region
print("Task 1 — Total Sales Per Region:")

regions= df['Region'].unique()
df_indexed = df.set_index('Region')

for region in regions:
    region_sales = df_indexed.loc[region, 'Sales']
    total        = np.sum(region_sales)
    print(f"  {region:<8}: Rs.{total:,}")

# Task 2: Identify the top salesperson
top_index       = df['Sales'].idxmax()
top_salesperson = df.loc[top_index, 'Salesperson']
top_sales       = df.loc[top_index, 'Sales']

print("Task 2 — Top Salesperson:")
print(f"  {top_salesperson} with Sales = Rs.{top_sales}")

# Task 3: Compute average sales across all regions
avg_sales = df['Sales'].mean()

print("Task 3 — Average Sales Across All Regions:")
print(f"  Average Sales: Rs.{avg_sales:}")

# Task 4: Sort the dataset by Sales in descending order
df_sorted = df.sort_values(by='Sales', ascending=False)
print("Task 4 — Dataset Sorted by Sales (Descending):")
print(df_sorted.to_string(index=False))
```

Q7. Customer Purchase Analysis
Dataset:
data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}
Tasks:
1. Find the average purchase amount by city.
2. Identify customers with purchase above overall average.
3. Find the city generating the highest revenue.
4. Create a column Age_Group:
o <30 → Young
o 30–50 → Adult
o 50 → Senior
5. Find average purchase per age group

```python
import numpy as np
import pandas as pd

# Customer purchase dataset
data = {
    'CustomerID'    : [101, 102, 103, 104, 105, 106, 107],
    'Age'           : [25, 45, 35, 50, 23, 40, 60],
    'PurchaseAmount': [250, 600, 400, 800, 150, 700, 900],
    'City'          : ['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}

df = pd.DataFrame(data)

df['Age'] = df['Age'].astype(int)
print(df.to_string(index=False))

# Task 1: Average purchase amount by city
print("Task 1 — Average Purchase Amount by City:")
cities= df['City'].unique()
df_indexed = df.set_index('City')

for city in cities:
    city_purchases = df_indexed.loc[city, 'PurchaseAmount']
    avg  = np.mean(city_purchases)
    print(f"  {city}: Rs.{avg}")

# Task 2: Customers with purchase above overall average
overall_avg  = df['PurchaseAmount'].mean()
above_avg_df = df[df['PurchaseAmount'] > overall_avg]
print(f"Task 2 — Customers Above Overall Average (Rs.{overall_avg}):")
print(above_avg_df.to_string(index=False))

# Task 3: City generating the highest total revenue
print("Task 3 — Total Revenue by City:")
city_revenues = {}
for city in cities:
    city_purchases = df_indexed.loc[city, 'PurchaseAmount']
    city_revenues[city] = np.sum(city_purchases)
    print(f"  {city:<10}: Rs.{city_revenues[city]:,}")

best_city = max(city_revenues, key=city_revenues.get)
print(f"\n  Highest Revenue City: {best_city} (Rs.{city_revenues[best_city]:,})")

# Task 4: Create Age_Group column
age = df['Age'].values.astype(int)

conditions = [
    age < 30,
    (age >= 30) & (age <= 50),
    age > 50
]
choices = ['Young', 'Adult', 'Senior']

df['Age_Group'] = np.select(conditions, choices, default='Unknown')
print("Task 4 — Dataset with Age_Group Column:")
print(df.to_string(index=False))

# Task 5: Average purchase per age group
print("Task 5 — Average Purchase per Age Group:")
age_groups     = df['Age_Group'].unique()
df_age_indexed = df.set_index('Age_Group')

for group in age_groups:
    group_purchases = df_age_indexed.loc[group, 'PurchaseAmount']
    avg = np.mean(group_purchases)
    print(f"  {group}: Rs.{avg}")
```

Q8. Time Series Sales Analysis
Generate data:
dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})
Tasks:
1. Convert Date to datetime index.
2. Compute monthly percentage growth.
3. Calculate 3-month rolling average sales.
4. Identify the month with highest growth rate

```python
import numpy as np
import pandas as pd

# Generate time series sales data
dates = pd.date_range('2024-01-01', periods=12, freq='ME')
sales = [200, 220, 250, 270, 300, 320, 350, 360, 380, 400, 420, 450]

df = pd.DataFrame({'Date': dates, 'Sales': sales})
print("Original Dataset:")
print(df.to_string(index=False))

# Task 1: Convert Date to datetime index

df = df.set_index('Date')
df.index = pd.to_datetime(df.index)
print("Task 1 — DataFrame with DateTime Index:")
print(df.to_string())

# Task 2: Compute monthly percentage growth

df['Growth_%'] = df['Sales'].pct_change() * 100

print("Task 2 — Monthly Percentage Growth:")
print(df[['Sales', 'Growth_%']].to_string())

# Task 3: Calculate 3-month rolling average sales

df['Rolling_Avg_3M'] = df['Sales'].rolling(window=3).mean()

print("Task 3 — 3-Month Rolling Average:")
print(df[['Sales', 'Rolling_Avg_3M']].to_string())

# Task 4: Identify the month with highest growth rate

best_month       = df['Growth_%'].idxmax()
best_growth_rate = df.loc[best_month, 'Growth_%']

print("Task 4 — Month with Highest Growth Rate:")
print(f"  Month       : {best_month.strftime('%B %Y')}")
print(f"  Growth Rate : {best_growth_rate}%")
```
